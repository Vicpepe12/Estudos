```python
import pandas as pd
```

Creating DataFrame from Dictionary

```python
data = {
    'Name': ['Victor', 'Denis', 'Cleber'],
    'Age': [20, 45, 1],
    'Score': [8.5, 3.0, 9.9]    
}
```

```python
pd.DataFrame(data)  
```

From List of Lists:

```python
data = [['Victor', 20, 8],
        ['Denis', 55, 3],
        ['Cleber', 2, 9]]
```

```python
df = pd.DataFrame(data, columns= ['Name', 'Age', 'Score'])
df
```
