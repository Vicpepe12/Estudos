```python
gauss_amarelo = pd.read_csv('./Dados_Freq/espectro_LED_Am.csv', sep='\s', decimal=',')
gauss_amarelo_y = np.array(gauss_amarelo['Int'])
gauss_amarelo_x = np.array(gauss_amarelo['Freq'])

mi_amarelo = gauss_amarelo_x[np.argmax(gauss_amarelo_y)]
print(mi_amarelo)
def func(x, sigma):
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(x-mi_amarelo)**2/(2*sigma**2))

popt, pcov = curve_fit(
    f = func,
    xdata = gauss_amarelo_x,
    ydata = gauss_amarelo_y
)

s_amarelo = popt

```

```python
freqs = [mi_amarelo, mi_azul, mi_verde, mi_violeta]
s_freqs = [s_amarelo[0], s_azul[0], s_verde[0], s_violeta[0]]
```

```python
V_0 = [amarelo['V_0'][0],azul['V_0'][0],verde['V_0'][0],violeta['V_0'][0]]
sV_0 = [amarelo['sV_0'][0],azul['sV_0'][0],verde['sV_0'][0],violeta['sV_0'][0]]
```

```python
e = -1.602176634e-19

V_0 = np.array(V_0)
E = V_0*e
sV_0 = np.array(sV_0)
s_E = -sV_0*e
def reta(x, h, phi):
    return h*x - e*phi

popt, pcov = curve_fit(
    f = reta,
    xdata = freqs,
    ydata = E,
    sigma = s_E,
    absolute_sigma = True
)

par = popt
perr = np.sqrt(np.diag(pcov))

x_model = np.linspace(min(freqs) - 0.1*min(freqs), max(freqs))  
y_model = reta(x_model, par[0], par[1])
par, perr
```

```python
s_E = [s_E[0]]*4
```

```python

graphWidth = 800
graphHeight = 600

f = plt.figure(figsize=(graphWidth/100.0, graphHeight/100.0), dpi=100)
grafico = f.add_subplot(111)
# plot dos pontos:
grafico3 = grafico.plot(freqs, E, marker = '.', markersize=5, color = 'black', linewidth = 0)
# grafico3 = grafico.plot(x_data_p, y_data_p, marker = '.', markersize=4, color = 'black', linewidth = 0)

# plot(dir_name_40, 'black')
# plot(dir_name_40, 'green')
# plot(dir_name_60, 'blue')
# plot(dir_name_100, 'red')


# plot do ajuste
grafico.plot(x_model, y_model,  color = 'red')

grafico.errorbar(freqs, E, yerr = s_E, ls='None', color = 'black')
# grafico.axvline(x = 0, color = 'grey')
# grafico.axhline(y = 0, color = 'grey')


grafico.grid(True)
grafico.set_xlabel('Frequência (Hz)') # eixo x
grafico.set_ylabel('Energia (eV)') # eixo y
grafico.set_title('Energia de corte x Frequência (Método 1)') #título
grafico.legend(loc = 'upper left', fontsize= 12)



#    plt.savefig('Acr25graus.png')
plt.show()
```
