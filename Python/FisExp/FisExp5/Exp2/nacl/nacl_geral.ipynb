```python
def gaussiana(x, sigma, mi):
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(x-mi)**2/(2*sigma**2))


def gaussiana_fit(dados_x, dados_y):
    popt, pcov = curve_fit(
    f = gaussiana,
    xdata = dados_x,
    ydata = dados_y
)
    return [popt, pcov]
```
